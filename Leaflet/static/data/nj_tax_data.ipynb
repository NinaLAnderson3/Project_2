```python
import pandas as pd
import csv
```

```python
data = pd.read_csv("nj_tax_data.csv", skiprows = [1,2,3])
```

```python
df = data.rename(columns = {"Unnamed: 0":"County", "Unnamed: 1" : "City_id", "Unnamed: 2": "City", "2020":"2020_tax_rate"})
```

```python
df.dropna()
```

```python
df.columns
```

```python
tax_df = df[["County","City_id","City","2020_tax_rate"]]
clean_tax = tax_df.dropna()
```

```python
clean_tax.rename(columns = {"2020_tax_rate":"tax_rate"})
```

```python
grouped_clean = clean_tax.groupby("County").mean()
grouped_clean.rename(columns = {"County":"county_name","2020_tax_rate":"Average_tax_rate"})
```

```python
loc_data = pd.read_csv("county_loc.csv")
loc_data.head()
```

```python
new_loc = loc_data.rename(columns = {"county_name":"County"})
```

```python
# data['Name'] = data['Name'].str.upper() 
new_loc["County"] = new_loc["County"].str.upper()
new_loc
```

```python
merge1 = pd.merge(grouped_clean, new_loc,on = "County", how = "left")
```

```python
merge1
```

```python
merge1.to_csv("tax_loc.csv")
```

```python
school_data = pd.read_csv("school_scores_2017_18.csv")
school_data.columns
```

```python
parsed_school = school_data[["COUNTY_CODE","COUNTY_NAME","SummativeRating","SummativeScore"]]
```

```python
new = parsed_school.rename(columns = {"COUNTY_NAME":"County"})
```

```python
grouped = new.groupby("County").mean()
```

```python
grouped
```

```python
# df.drop('Charlie', axis=0))
grouped.drop("CHARTERS", axis =0)
```

```python
full_df = pd.merge(merge1,grouped,on="County",how = "left")
```

```python
full_df.to_csv("final.csv")
```
