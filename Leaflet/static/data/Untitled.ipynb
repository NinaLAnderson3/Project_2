```python
import pandas as pd
import json
import csv
```

```python
data = pd.read_csv("nj_tax_data.csv", skiprows = [1,2,3])
updated_df = data.rename(columns = {"Unnamed: 0":"County","2020":"tax_rate_2020"})
updated_df.dropna()
```

```python
new_df = updated_df[["County","tax_rate_2020"]]
new_df.dropna()
```

```python
grouped_county_tax = new_df.groupby("County").mean()["tax_rate_2020"]
```

```python
grouped_county_tax
```

```python
cities = pd.read_csv("cities.csv",encoding='latin-1')
```

```python
nj_df = cities.loc[(cities.state_name == "New Jersey")]
nj_df.columns
```

```python
grouped_nj = nj_df.groupby("county_name").mean()
grouped_nj
```

```python
parsed_nj = grouped_nj[["lat","lng"]]
parsed_nj
```

```python
location = []
for key, row in parsed_nj.iterrows():
    locs = []
    locs.append(row["lat"])
    locs.append(row["lng"])
    location.append(locs)

  
```

```python
location
```

```python
parsed_nj['location'] = location
```

```python
parsed_nj.to_csv("nj_county_loc.csv")
```

```python
parsed_nj.head()
```

```python
crime_df = pd.read_csv("nj_crime_detail.csv")
```

```python
crime_df.head()
```

```python
renamed_crime = crime_df.rename(columns = {"county_name":"County"})
```

```python
clean_crime = renamed_crime.drop(["police_dept","crime_type"], axis = 1)
```

```python
clean_crime.head()
```

```python
grouped_crime = clean_crime.groupby(["County"]).mean()
grouped_crime.head()
```

```python
merged_nj = pd.merge(grouped_crime, clean_final, on = "County", how = "left")
```

```python
merged_nj
```

```python
final_nj = merged_nj.rename(columns = {"count":"Crime_Count"}).drop(["Unnamed: 0"], axis = 1)
```
